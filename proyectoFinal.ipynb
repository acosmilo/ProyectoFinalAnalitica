```python
from statsmodels.stats.diagnostic import normal_ad

# Supongamos que tienes un dataframe df y la columna de interés se llama 'columna'
columna = dfc['Avg_glucose_level']

# Realizar prueba de normalidad de Anderson-Darling
statistic, pvalue = normal_ad(columna)

# Acceder a los resultados
print('Estadístico de prueba:', statistic)
print('Valor p:', pvalue)

# Interpretación
if pvalue > 0.05:
    print('La columna sigue una distribución normal (hipótesis nula no se rechaza)')
else:
    print('La columna no sigue una distribución normal (hipótesis nula se rechaza)')
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
from scipy.stats import ttest_rel


# Datos de ejemplo
X1 = X_balance2
y1 = y_balance2
X2 = X_balance3
y2 = y_balance3

# Realizar validación cruzada en ambos modelos
n_splits = 6
scores_modelo1 = cross_val_score(model1, X1, y1, cv=n_splits, scoring='accuracy') # Cambiar 'accuracy' por la métrica deseada
scores_modelo2 = cross_val_score(model1, X2, y2, cv=n_splits, scoring='accuracy') # Cambiar 'accuracy' por la métrica deseada

# Realizar prueba de hipótesis (t-test)
t_stat, p_value = ttest_rel(scores_modelo1, scores_modelo2)

# Imprimir resultados
print("Modelo 1: Accuracy medio = {:.3f}, Desviación estándar = {:.3f}".format(np.mean(scores_modelo1), np.std(scores_modelo1)))
print("Modelo 2: Accuracy medio = {:.3f}, Desviación estándar = {:.3f}".format(np.mean(scores_modelo2), np.std(scores_modelo2)))
print("Resultado del t-test: Estadístico t = {:.3f}, Valor p = {:.3f}".format(t_stat, p_value))

#  Hipótesis nula (H0): No hay diferencia significativa entre las medias de las dos muestras relacionadas.
#  Hipótesis alternativa (H1): Hay una diferencia significativa entre las medias de las dos muestras relacionadas.
        
        
```

#### learning_rate{‘constant’, ‘invscaling’, ‘adaptive’}, default=’constant’

Tasa de aprendizaje (learning rate): determina la magnitud del ajuste de los pesos en cada actualización. Si es demasiado alta, puede causar oscilaciones y divergencia, mientras que si es demasiado baja, puede tardar mucho tiempo en converger o nunca converger.

    Learning rate schedule for weight updates.

        ‘constant’ is a constant learning rate given by ‘learning_rate_init’.

        ‘invscaling’ gradually decreases the learning rate at each time step ‘t’ using an inverse scaling exponent of ‘power_t’. effective_learning_rate = learning_rate_init / pow(t, power_t)

        ‘adaptive’ keeps the learning rate constant to ‘learning_rate_init’ as long as training loss keeps decreasing. Each time two consecutive epochs fail to decrease training loss by at least tol, or fail to increase validation score by at least tol if ‘early_stopping’ is on, the current learning rate is divided by 5.
